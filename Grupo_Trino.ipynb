```python
import pandas as pd
```

```python
df = pd.read_excel('/content/survey-data-trino-cx.xlsx')
```

```python
df
```

```python
df.info()
```

Os valores que não são numéricos retorna como nulo

```python
# Converta a coluna para numérico
df['Prazo'] = pd.to_numeric(df['Prazo'], errors='coerce')
```

```python
df[['Prazo']]
```

Pegar os valores que são negativos e preencher como valor 1

```python
df.iloc[:, 8:] = df.iloc[:, 8:].applymap(lambda x: 1 if x < 0 else x)
```

Verificando se a coluna que tinha valores negativos converteram pra 1

```python
df[['Próximo']]
```

Pegar os valores nulos das colunas numéricas e colocar como a média e aproximar pra o numero inteiro mais próximo

```python
colunas = df.columns[8:]

# Preencher os valores nulos de cada coluna com a média
df[colunas] = df[colunas].fillna(df[colunas].mean())
```

```python
df[['Encantado']]
```

Transformar todas as colunas do indice 8 em diante em valores inteiros

```python
df.iloc[:, 8:] = df.iloc[:, 8:].astype(int)
```

Importando o dataset modificado do colab pro computador

```python
df.to_csv('trino_modificado.csv', index = False)
```
