```python
import pandas as pd
import numpy as np
```

```python
# ler o arquivo CSV (supondo que o separador é a vírgula e o arquivo não tem cabeçalho)
df = pd.read_excel('/content/trino_excel.xlsx')
```

```python
df
```

```python
df.info()
```

```python
# Selecione as colunas de 9 em diante que possuem Dtype igual a 'object'
cols_obj = df.iloc[:, 9:].select_dtypes(include='object').columns
```

```python
# Transforme as linhas que não são numéricas em valores nulos
df[cols_obj] = df[cols_obj].applymap(lambda x: pd.to_numeric(x, errors='coerce'))
```

```python
df['Vagas']
```

```python
# Selecionando as colunas a partir da coluna 9 
cols_to_fill = df.columns[9:]
```

```python
# Calculando a média das colunas selecionadas
mean = df[cols_to_fill].mean()
```

```python
# Preenchendo os valores nulos nas colunas selecionadas com a média das colunas
df[cols_to_fill] = df[cols_to_fill].fillna(mean)
```

```python
df
```

```python
# iterando sobre as colunas relevantes e arredondando valores para inteiros
for i in range(9, len(df.columns)):
    df.iloc[:, i] = df.iloc[:, i].apply(lambda x: round(x))
```
